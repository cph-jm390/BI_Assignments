```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread


def readJPG(imagePath):
# Read the image
    image_data = [imread(imagePath)]

#  DataFrame
    W = {'img': image_data}
    df = pd.DataFrame(W)

# Display the image
    plt.imshow(df.img[0])
    plt.show()
    


```

```python
## the first function readMPG is for windows, DOES NOT WORK ON MAC

import cv2
def readMPG(videoPath):
    
    cap = cv2.VideoCapture(videoPath)

    while True:
        ret, frame = cap.read()
        
        # If the frame was not read successfully, break the loop
        if not ret:
            print('corrupted frame')
            break
        
        # Display the frame each frame in the loop for a video
        cv2.imshow('Video', frame)
        
        # press q for quit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
    
# The following function which is outcommented is for mac
#import subprocess

#def openWithQuickTime(videoPath):
#    try:
#       subprocess.run(['open', '-a', 'QuickTime Player', videoPath])
        
#    except Exception as e:
#        print(f"An error occurred: {e}")

```

```python
import wikipediaapi

def readWiki(pagePath):
    # Create a Wikipedia API object
    wiki_wiki = wikipediaapi.Wikipedia('en', extract_format=wikipediaapi.ExtractFormat.WIKI, headers={'User-Agent': 'Your_User_Agent_Here'})    
    # Replace 'Page_Title' with the title of the Wikipedia page you want to read
    page = wiki_wiki.page(pagePath)
    
    # Check if the page exists
    if page.exists():
        print("Title:", page.title)
        print("Summary:", page.summary)
    else:
        print("Page does not exist.")
    
```
